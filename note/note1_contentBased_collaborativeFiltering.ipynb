### **2. C√°ch hai ph∆∞∆°ng ph√°p ƒë∆∞·ª£c k·∫øt h·ª£p trong notebook n√†y**

#### **Ph∆∞∆°ng ph√°p 1: Content-Based Filtering**
- **S·ª≠ d·ª•ng TF-IDF (Term Frequency-Inverse Document Frequency)** ƒë·ªÉ chuy·ªÉn ƒë·ªïi n·ªôi dung c·ªßa c√°c s·∫£n ph·∫©m th√†nh c√°c ƒë·∫∑c tr∆∞ng s·ªë.

- H√†m ƒë∆∞·ª£c s·ª≠ d·ª•ng:

  tfidf_vectorizer = TfidfVectorizer()

  def get_content_based_recommendations(product_id, top_n):
