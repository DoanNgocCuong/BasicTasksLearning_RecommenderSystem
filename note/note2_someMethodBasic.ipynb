## ĐÁNH GIÁ CÁC METHOD với BỘ DATASET TRONG BÀI 

### Dataset này sẽ phù hợp với những phương pháp nào và tại sao?

Dựa trên đặc điểm của dataset từ **OTTO Recommender System**, bao gồm lịch sử hành vi người dùng trong các phiên (sessions) và các loại hành động như **clicks**, **carts**, và **orders**, dưới đây là các phương pháp phù hợp:

---

### **1. Base Method: Most Popular (Phổ biến nhất)**
- **Cách hoạt động:**
  - Tính tần suất xuất hiện của các sản phẩm trong toàn bộ dữ liệu (train).
  - Gợi ý các sản phẩm được nhấp chuột, thêm vào giỏ hàng, hoặc đặt mua nhiều nhất.
  
- **Tại sao phù hợp:**
  - Dataset có **220 triệu sự kiện** và hơn **1.8 triệu sản phẩm** -> dữ liệu phong phú giúp phân loại các sản phẩm phổ biến.
  - Không phụ thuộc vào lịch sử của người dùng -> dễ triển khai, hiệu quả nhanh.

- **Hạn chế:**
  - Thiếu tính cá nhân hóa -> mọi người dùng sẽ nhận được danh sách gợi ý giống nhau.

---

### **2. Session-Based Collaborative Filtering**
- **Cách hoạt động:**
  - Tìm các phiên làm việc (sessions) có hành vi tương tự (dựa trên sản phẩm đã click/thêm/mua).
  - Gợi ý sản phẩm từ các phiên tương tự.

- **Tại sao phù hợp:**
  - Dataset có cấu trúc session rõ ràng, cho phép tìm các phiên tương đồng.
  - Phù hợp với bài toán **session-based recommendation**, nơi các phiên ngắn không cần liên kết lâu dài giữa các người dùng.

- **Hạn chế:**
  - Không hoạt động tốt khi các phiên có ít hành động (dữ liệu sparse).
  - Khó mở rộng với dữ liệu rất lớn.

---

### **3. Content-Based Filtering**
- **Cách hoạt động:**
  - Dựa trên đặc điểm sản phẩm (ví dụ: danh mục, thương hiệu) để gợi ý sản phẩm tương tự với những sản phẩm mà người dùng đã tương tác.

- **Tại sao phù hợp:**
  - Nếu dataset bổ sung thông tin metadata về sản phẩm (e.g., thể loại, giá), phương pháp này sẽ tận dụng được để gợi ý.
  - Phù hợp khi sản phẩm mới xuất hiện mà chưa có nhiều dữ liệu tương tác.

- **Hạn chế:**
  - Dataset hiện tại không chứa metadata về sản phẩm -> cần thêm dữ liệu bổ sung.
  - Thiếu khả năng tận dụng hành vi người dùng khác (cộng đồng).

---

### **4. Matrix Factorization (Collaborative Filtering bằng Latent Factors)**
- **Cách hoạt động:**
  - Sử dụng thuật toán như **SVD**, **ALS** để phân tích ma trận người dùng-sản phẩm hoặc session-sản phẩm, tìm ra các "yếu tố ẩn" chung giữa các sản phẩm/người dùng.

- **Tại sao phù hợp:**
  - Với dataset lớn, matrix factorization khai thác mối quan hệ giữa các phiên và sản phẩm hiệu quả hơn so với phương pháp dựa trên tần suất.
  - Phù hợp với bài toán đa mục tiêu (clicks, carts, orders), khi kết hợp các loại hành động vào một ma trận.

- **Hạn chế:**
  - Không giải quyết tốt bài toán cold-start (khi sản phẩm hoặc người dùng mới không có dữ liệu tương tác).

---

### **5. Neural Networks (Deep Learning for Recommendations)**
- **Cách hoạt động:**
  - Mô hình hóa hành vi người dùng trong các phiên với mạng nơ-ron, ví dụ:
    - **Recurrent Neural Networks (RNNs):** Mô hình hóa tuần tự các hành động.
    - **Transformers:** Hiểu ngữ cảnh trong các hành động theo cách phi tuần tự.

- **Tại sao phù hợp:**
  - Dataset có cấu trúc tuần tự rõ ràng trong từng session -> RNNs hoặc Transformers có thể tận dụng.
  - Dataset đủ lớn để huấn luyện các mô hình deep learning mà không lo vấn đề overfitting.

- **Hạn chế:**
  - Yêu cầu tài nguyên tính toán lớn.
  - Cần thiết kế mô hình cẩn thận để xử lý đa mục tiêu (clicks, carts, orders).

---

### **6. Hybrid Methods (Kết hợp Content-Based và Collaborative Filtering)**
- **Cách hoạt động:**
  - Kết hợp Collaborative Filtering và Content-Based Filtering để tận dụng cả lịch sử hành vi người dùng và thông tin metadata sản phẩm.

- **Tại sao phù hợp:**
  - Nếu bổ sung metadata, Hybrid Methods có thể gợi ý chính xác hơn nhờ khai thác cả nội dung sản phẩm và hành vi cộng đồng.
  - Dataset có nhiều phiên ngắn -> kết hợp cả session-based và content-based để cải thiện độ chính xác.

- **Hạn chế:**
  - Phức tạp khi triển khai và yêu cầu xử lý dữ liệu đa chiều.

---

### **7. Graph-Based Methods (Sử dụng đồ thị để gợi ý)**
- **Cách hoạt động:**
  - Xây dựng đồ thị với các nút là sản phẩm và các cạnh đại diện cho tương tác của người dùng (clicks, carts, orders).
  - Áp dụng thuật toán như **Node2Vec**, **Graph Neural Networks (GNNs)** để gợi ý.

- **Tại sao phù hợp:**
  - Dataset chứa các loại hành động liên kết chặt chẽ giữa sản phẩm -> đồ thị giúp khai thác mối quan hệ sâu sắc hơn giữa các sản phẩm.
  - Phù hợp với các sản phẩm ít tương tác (cold-start).

- **Hạn chế:**
  - Cần kỹ thuật phức tạp và tài nguyên tính toán lớn.

---

- **Base method:** Most Popular -> để thiết lập baseline.

### **Đặc điểm của dataset:**
1. **Session-based**:
   - Các hành vi được tổ chức theo phiên làm việc (sessions).
   - Mỗi session có một loạt các sự kiện, như **clicks**, **carts**, và **orders**.
   - Không có thông tin về người dùng cụ thể -> không thể áp dụng user-based recommendation.

2. **Hành động đa mục tiêu (multi-target)**:
   - Ba loại hành động cần dự đoán: clicks, carts, orders.
   - Phân phối hành động có thể không đồng đều (clicks nhiều hơn orders).

3. **Dữ liệu lớn (Big Data)**:
   - 220 triệu sự kiện, hơn 1.8 triệu sản phẩm.
   - Đòi hỏi phương pháp có khả năng xử lý hiệu quả với dữ liệu lớn.

4. **Dữ liệu tuần tự**:
   - Các hành động trong một session được sắp xếp theo thứ tự thời gian.
   - Tính tuần tự có ý nghĩa quan trọng để dự đoán sản phẩm tiếp theo.
